```python
pokemon["Name_nospace"] = pokemon["Name"].apply(lambda i: i.replace(" ", ""))
pokemon["name_isalpha"] = pokemon["Name_nospace"].apply(lambda i: i.isalpha())
pokemon[pokemon["name_isalpha"] == False]
```

* 모든 이름이 알파벳으로만 이루어졌군!

```python
import re
```

```python
def tokenize(name):
    name_split = name.split(" ")
    
    tokens = []
    for part_name in name_split:
        a = re.findall('[A-Z][a-z]*', part_name)
        tokens.extend(a)
        
    return np.array(tokens)
```

```python
all_tokens = list(legendary["Name"].apply(tokenize).values)

token_set = []
for token in all_tokens:
    token_set.extend(token)

print(len(set(token_set)))
print(token_set)
```

```python
from collections import Counter
```

```python
most_common = Counter(token_set).most_common(10)
most_common
```

```python
# 토크나이징한 이름 컬럼 생성
for token, _ in most_common:
    pokemon[token] = pokemon["Name"].str.contains(token)

pokemon.head(10)
```

## 머신러닝 베이스라인 만들기

베이스라인 : 가장 기초적인 데이터로 만든 모델.
* 성능 하한선을 제공함으로써 만든 모델이 적절한지 판단할 수 있다.

```python
print(original_data.shape)
original_data.head()
```

```python
original_data.columns
```

```python
# 컬럼 중 의미있는 데이터만 추림.
features = ['Total', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']
```

```python
target = 'Legendary'
```

```python
X = original_data[features]
print(X.shape)
X.head()
```

```python
y = original_data[target]
print(y.shape)
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15)

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(random_state=25)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

* Positive : 전설의 포켓몬 Negative : 일반 포켓몬
* legendary 포켓몬은 65마리, 일반 포켓몬은 약 800마리이므로, 해당 데이터는 불균형 데이터이다.
* recall이 낮음(FN값이 높음). 즉, 전설의 포켓몬인데 일반 포켓몬으로 판단하는 경우가 많다.
* Positive 데이터가 적으므로, 적은 양의 데이터인 전설의 포켓몬을 잘 잡아내는 것이 중요하다.
